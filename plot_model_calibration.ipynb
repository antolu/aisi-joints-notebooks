Notebook to visualize the effect of model calibration on Inception ResNet V2

```python
calibrated_test_csv = 'evaluated_test_calibrated.csv'
uncalibrated_test_csv = 'evaluated_test_uncalibrated.csv'

calibrated_validation_csv = 'evaluated_validation_calibrated.csv'
uncalibrated_validation_csv = 'evaluated_validation_uncalibrated.csv'

KTH_BLUE = '#0073FF'
N_BINS = 20
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import reliability_diagrams as rd
```

```python
uncalibrated_test = pd.read_csv(uncalibrated_test_csv)
calibrated_test = pd.read_csv(calibrated_test_csv)

uncalibrated_val = pd.read_csv(uncalibrated_validation_csv)
calibrated_val = pd.read_csv(calibrated_validation_csv)
```

```python
fig = rd.reliability_diagram(uncalibrated_test['cls'], uncalibrated_test['detected_class'], confidences=uncalibrated_test['detection_score'], num_bins=N_BINS, title='Reliability Diagram, Uncalibrated Inception ResNet V2, test set', return_fig=True)
fig.savefig('inception_resnet_uncalibrated_test.pdf', format='pdf')

fig = rd.reliability_diagram(calibrated_test['cls'], calibrated_test['detected_class'], calibrated_test['detection_score'], num_bins=N_BINS, title='Reliability Diagram, Calibrated Inception ResNet V2, test set', return_fig=True)
fig.savefig('inception_resnet_calibrated_test.pdf', format='pdf')

fig = rd.reliability_diagram(uncalibrated_val['cls'], uncalibrated_val['detected_class'], confidences=uncalibrated_val['detection_score'], num_bins=N_BINS, title='Reliability Diagram, Uncalibrated Inception ResNet V2, validation set', return_fig=True)
fig.savefig('inception_resnet_uncalibrated_val.pdf', format='pdf')

fig = rd.reliability_diagram(calibrated_val['cls'], calibrated_val['detected_class'], calibrated_val['detection_score'], num_bins=N_BINS, title='Reliability Diagram, Calibrated Inception ResNet V2, validation set', return_fig=True)
fig.savefig('inception_resnet_calibrated_val.pdf', format='pdf')
```
