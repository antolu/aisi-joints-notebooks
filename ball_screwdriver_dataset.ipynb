```python
import pandas as pd
import os
from os import path
import uuid
import sklearn.model_selection

from aisi_joints.img_cls import train
```

```python
workspace = '/daten/results-work-dir/research/20220309_Schienenstosse_master/data/anton/ball_screwdriver'
dataset_root = path.join(workspace, 'training_data')

csv_path = path.join(dataset_root, 'dataset.csv')

os.chdir(workspace)

train_split = 0.8
```

```python
samples_ok = []
samples_def = []


for file in os.listdir(dataset_root):
    sample = {
        'eventId': uuid.uuid1(),
        'filepath': path.abspath(path.join(dataset_root, file)),
        'x0': 0,
        'x1': 149,
        'y0': 0,
        'y1': 149,
    }
    if file.startswith('P'):
        sample['cls'] = 'DEFECT'
        samples_def.append(sample)
    elif file.startswith('N'):
        sample['cls'] = 'OK'
        samples_ok.append(sample)
    elif file == path.split(csv_path)[1]:
        continue
    else:
        raise ValueError(f'File {file} does not conform to a recognized format')
        
train_ok, val_ok = sklearn.model_selection.train_test_split(samples_ok, train_size=train_split)
train_def, val_def = sklearn.model_selection.train_test_split(samples_def, train_size=train_split)

for elem in train_ok + train_def:
    elem['split'] = 'train'
for elem in val_ok + val_def:
    elem['split'] = 'validation'

dataset = pd.DataFrame(train_ok + train_def + val_ok + val_def)

dataset.to_csv(csv_path, index=False)
```

```python
train_args = ['configs/config.py', '-d', csv_path]

train.main(train_args)
```
